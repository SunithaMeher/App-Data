```python
#defining a function to view some more insights of the dataset.
def printinfo():
    temp = pd.DataFrame(index=data.columns)
    temp['data_type'] = data.dtypes
    temp['null_count'] = data.isnull().sum()
    temp['unique_count'] = data.nunique()
    return temp
printinfo()
```

Replacing the missing values of the Rating Column with the Mode value of that entire column.

```python
modeValueRating = data['Rating'].mode()
modeValueRating[0]
```

```python
data['Rating'].fillna(value=modeValueRating[0], inplace = True)
printinfo()
```

Thus,completed the data cleaning process.

Before continuing with data preparation the columns 'Reviews','Size','Installs',&'Price' should be converted from object type  to integer or float datatype.

```python
#Converting the Reviews column to integer.
data['Reviews'] = data.Reviews.astype(int)
printinfo()
```

```python
#Converting the Size Column from object to integer:
```

```python
data['Size'] = data.Size.apply(lambda x: x.strip('+'))#removing the '+'symbol
```

```python
data['Size'] = data.Size.apply(lambda x: x.replace(',', ''))#removing the `,`symbol
```

```python
data['Size'] = data.Size.apply(lambda x: x.replace('M', 'e+6'))# For converting the M to Mega 
#Replacing the M by multiplying the value with 1000000
```

```python
data['Size'] = data.Size.apply(lambda x: x.replace('k', 'e+3'))# For convertinf the K to Kilo
#Replacing the k by multiplying the value with 1000
```

```python
data['Size'] = data.Size.replace('Varies with device', np.NaN)
#Replacing the Varies with device value with Nan.
```

```python
data['Size'] = pd.to_numeric(data['Size']) # Converting the string to Numeric type
printinfo()
```

```python
#dropping Rows of the column Size having Nan values
data.dropna(subset = ['Size'], inplace=True)
printinfo()
```

```python
#Converting the Installs column from object to integer
```

```python
data['Installs'] = data.Installs.apply(lambda x: x.strip('+')) #removing the '+'symbol
```

```python
data['Installs'] = data.Installs.apply(lambda x: x.replace(',', ''))#removing the `,`symbol
```

```python
data['Installs'] = pd.to_numeric(data['Installs'])#converting it from string type to numeric type
printinfo()
```

```python
#Converting the Price column from object to numeric
```

```python
data['Price'].value_counts()
```

```python
data['Price'] = data.Price.apply(lambda x: x.strip('$')) #removing the '$'symbol
```

```python
data['Price'] = pd.to_numeric(data['Price'])#converting the values to Numeric type.
printinfo()
```

Visualizing the plot:

```python
y = data['Category'].value_counts().index
x = data['Category'].value_counts()
xasis = []
yasis = []
for i in range(len(x)):
    xasis.append(x[i])
    yasis.append(y[i])
```

```python
plt.figure(figsize=(25,15))
plt.xlabel("Count")
plt.ylabel("Category")
graph = sns.barplot(x = xasis, y = yasis, palette= "husl")
graph.set_title("Top categories in the Android Market", fontsize =24);
```
